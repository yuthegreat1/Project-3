```python
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.pro-football-reference.com/years/2021/receiving.htm'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

table = soup.find('table', class_='per_match_toggle')

players = table.find_all('tr')
print(players)

    
browser.quit()
```

```python
len(players)
```

```python
names=[]
stats=[]
for player in players:
    name=""
    position=""
    targets=""
    catches=""
    odds=""
    yards=""
    tds=""
    td = player.find('td')
    if td:
        if "*" in td.text:
            names.append(td.text[0:td.text.find("*")])
        names.append(td.text)
        name= td.text[0:td.text.find("*")]
    td = player.find('td', attrs={'data-stat': 'pos'}) 
    if td:
        position=td.text
    td = player.find('td', attrs={'data-stat': 'targets'})
    if td:
        targets=td.text
        td = player.find('td', attrs={'data-stat': 'rec'})
    if td:
        catches=td.text
        td = player.find('td', attrs={'data-stat': 'catch_pct'})
    if td:
        odds=td.text
        td = player.find('td', attrs={'data-stat': 'rec_yds'})
    if td:
        yards=td.text
        td = player.find('td', attrs={'data-stat': 'rec_td'})
    if td:
        tds=td.text
    stat={
        'name':name,
        'position': position,
        'targets': targets,
        'receptions': catches,
        'catch percentage': odds,
        'yards': yards,
        'touchdowns': tds
    }
    stats.append(stat)
```
